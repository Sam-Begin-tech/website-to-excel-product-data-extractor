````python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

visited = set()

def crawl(url, domain):
    if url in visited or domain not in url:
        return []
    visited.add(url)
    print(f"Crawling: {url}")
    
    try:
        response = requests.get(url, timeout=10)
        soup = BeautifulSoup(response.text, "html.parser")
        texts = [soup.get_text()]
        for link in soup.find_all("a", href=True):
            next_url = urljoin(url, link['href'])
            if domain in next_url:
                texts += crawl(next_url, domain)
        return texts
    except:
        return []
start_url = "https://edenmart.in/"
domain = urlparse(start_url).netloc
all_texts = crawl(start_url, domain)
all_texts

````

````python
import re

def clean_text(raw_text):
    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', raw_text)
    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'not found 404 not found', '', text)  # Remove 404 errors
    text = re.sub(r'loading\.\.\.|add to cart|shop now|my cart|login|register', '', text)

    return text.strip()

new_texts = ""
for raw_text in all_texts:
    if raw_text:
        new_texts += clean_text(raw_text) + "\n---\n"
     
print(new_texts)

````

````python
def chunk_text(text, max_chars=2000):
    return [text[i:i+max_chars] for i in range(0, len(text), max_chars)]
chunks=chunk_text(new_texts)
chunks
````

````python
from langchain_ollama import OllamaLLM

llm = OllamaLLM(model="gemma2:2b",temperature=0)

llm.invoke("What is the dollar based gross retention rate:")
````

````python
replies = []
for chunk in chunks:
    prompt = f"""
Extract the following fields from the text: Product Name, MRP, and Selling Price.

Only respond with a **valid JSON object** in the exact format below, using actual values **from the text**:
{{
    "ProductName": "Exact product name from the text",
    "MRP": "₹[amount]",      # e.g., ₹310
    "sellingPrice": "₹[amount]"  # e.g., ₹300
}}

- Do NOT use placeholders like ₹xxx or "Product name".
- If **any value is missing or unclear**, return **null** for that field (not empty string or placeholder).
- Do not explain anything or include any extra text—just the JSON object.

Text:
\"\"\"{chunk}\"\"\"
"""
    repy = llm.invoke(prompt)
    replies.append(repy)
replies
````

````python
import json
import re

def clean_json_string(raw_output: str) -> str:
    """Remove markdown backticks and extra formatting."""
    return re.sub(r"```(json)?", "", raw_output).strip("` \n")



data = []
for item in replies:
    cleaned = clean_json_string(item)
    try:
        parsed = json.loads(cleaned)
        data.append(parsed)
    except json.JSONDecodeError as e:
        print("Error parsing JSON:", e)
        print("Problematic string:", cleaned)

print(data)
````

````python
data
````

````python
unique_data = {}

for item in data:
    if not isinstance(item, dict):
        continue  # Skip if item is None or not a dict

    product_name = item.get("ProductName", "")
    if not isinstance(product_name, str) or not product_name.strip():
        continue  # Skip if ProductName is not a string or is empty

    product_name = product_name.strip()

    # Keep only the first occurrence
    if product_name not in unique_data:
        unique_data[product_name] = item

# Convert back to a list
deduplicated_list = list(unique_data.values())

print(deduplicated_list)
````

````python
import pandas as pd
import numpy as np

# Convert to DataFrame
df = pd.DataFrame(deduplicated_list)

# Replace common null-like values with actual NaN
df.replace(["null", "None","", None], np.nan, inplace=True)

# Now drop rows where 'ProductName' is NaN
df = df.dropna(subset=['ProductName'])

# Write to Excel
df.to_excel("productsnew4.xlsx", index=False)

df
````
